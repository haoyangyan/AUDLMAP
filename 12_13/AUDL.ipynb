```python
# read the data
f = open(r"AUDLallhomeev.txt","r")
allhomeev = eval(f.read())
f = open(r"AUDLallawayev.txt","r")
allawayev = eval(f.read())
```

```python
allev = allhomeev+allawayev
```

```python
passf = ['thrower','throwerX','throwerY','receiver','receiverX','receiverY']
passf += ['team','opponent','gameid','scoreinpo','pval','pnext','timeout','evid','evtype']
passf += ['TXid','TYid','RXid','RYid','year','comp','X','Y','zoneid']
allpass = pd.DataFrame([],columns=passf)
for i in range(len(allev)):
    ev = allev[i]
    p = []
    for j in range(len(ev)):
        if ev[j]['type'] in [18,19,20,22,23]:
            if ev[j]['thrower'] != 'unknown':
                pj = [0]*len(passf)
                if ev[j]['type'] in [18,19,20]:
                    for k in range(6):
                        pj[k] = ev[j][passf[k]]
                if ev[j]['type'] in [22,23]:
                    for k in range(3):
                        pj[k] = ev[j][passf[k]]
                    pj[4] = ev[j]['turnoverX']
                    pj[5] = ev[j]['turnoverY']
                if i < len(games):
                    pj[6] = games[i]['homeTeamID']
                    pj[7] = games[i]['awayTeamID']
                    pj[8] = games[i]['gameID']
                else:
                    pj[6] = games[i-len(games)]['awayTeamID']
                    pj[7] = games[i-len(games)]['homeTeamID']
                    pj[8] = games[i-len(games)]['gameID']
                for k in range(j,len(ev)):
                    temp = ev[k]['type']
                    if temp == 19:
                        pj[9] = 1
                        break
                    if temp in [20,22,23,24] or temp in range(28,34):
                        pj[9] = 0
                        break
                for k in range(j,-1,-1):
                    if ev[k]['type'] in [1,2]:
                        pj[10] = k
                        break
                for k in range(j,len(ev)):
                    if ev[k]['type'] in [1,2]:
                        pj[11] = k
                        break
                for k in range(j,pj[10],-1):
                    if ev[k]['type'] in [3,5]:
                        pj[12] = k
                        break
                pj[13] = j
                pj[14] = ev[j]['type']
                pj[15] = int((pj[1]+27.5)/5)*5
                pj[16] = int((pj[2]-0.01)/5)*5
                pj[17] = int((pj[4]+27.5)/5)*5
                pj[18] = int((pj[5]-0.01)/5)*5
                pj[19] = int(str.split(pj[8],'-')[0])
                if pj[14] in [18,19]:
                    pj[20] = 1
                else:
                    pj[20] = 0
                pj[21] = round(pj[4] - pj[1],2)
                pj[22] = round(pj[5] - pj[2],2)
                radius = (pj[21]**2+pj[22]**2)**(1/2)
                if radius <= 5:
                    if pj[22]/abs(pj[21]+0.001) >= 1:
                        pj[23] = 1
                    else:
                        pj[23] = 2
                elif radius <= 30:
                    if pj[22]/abs(pj[21]+0.001) >= 1:
                        if pj[21] >= 0:
                            pj[23] = 3
                        else:
                            pj[23] = 4
                    else:
                        if pj[21] >= 0:
                            pj[23] = 6
                        else:
                            pj[23] = 5
                else:
                    if pj[22]/abs(pj[21]+0.001) >= 1:
                        pj[23] = 7
                    else:
                        pj[23] = 8
                p += [pj]
    passing = pd.DataFrame(np.array(p),columns=passf)
    allpass = pd.concat([allpass,passing],ignore_index=True)
```
